```python
def clean_data(df):
    df = df.T
    df = df.reset_index()
    df.columns = df.iloc[0]
    df = df.drop(df.index[0])
    # renaming the coloumn Profit & Loss account of Axis Bank (in Rs. Cr.) to sector
    df = df.rename(columns={'Profit & Loss account of Axis Bank (in Rs. Cr.)':'year'})
    # removing the rows with NaN values in Income from investment column
    # removing the rows with NaN values in Income from investment column
    df = df.dropna(subset=['Income from Investments'])
    # removing the coloumns with NaN values
    df = df.dropna(axis=1)
    # removing the coloumns with '--' values
    # it keeps only the rows with no '--' values in them
    df = df.loc[:, (df != '--').all(axis=0)]
    # remove a coloum if all values are 0 in it
    df = df.loc[:, (df != '0').any(axis=0)]
    for i in to_remove:
        # check if the coloumn is present in the dataframe
        if i in df.columns:
            df = df.drop(i, axis=1)

    df = df.apply(lambda x: x.str.replace(',', '') if x.dtype == 'object' else x)
    df = df.apply(pd.to_numeric, errors='ignore')
    # removing the object type coloumns
    df = df.select_dtypes(exclude=['object'])
    # removing the duplicate coloumns
    df = df.loc[:, ~df.columns.duplicated()]
    # to remove coloumns with objects in it
    list = []
    for col in df.columns:
        if df[col].dtype == 'object':
            list.append(col)
    df = df.drop(list[1:], axis=1)
    return df
```

```python
def normalize(df):
    # normalizing INCOME Interest / Discount on Advances / Bills, Income from Investments, Interest on Balance with RBI and Other Inter-Bank funds, Others w.r.t Total assets
    income_cols = ['Interest / Discount on Advances / Bills', 'Income from Investments', 'Interest on Balance with RBI and Other Inter-Bank funds', 'Others', 'Total Income', 'Other Income']
    for i in income_cols:
         if i in df.columns:
            df[i] = df[i] / df['Total Assets'] * 100

    # normailzing expenditure Interest Expended, Payments to and Provisions for Employees, Depreciation, Operating Expenses (excludes Employee Cost & Depreciation),
    # Total Provisions and Contingencies, Total Expenditure wrt Total Assets
    expenditure_cols = ['Interest Expended', 'Payments to and Provisions for Employees', 'Depreciation', 'Operating Expenses (excludes Employee Cost & Depreciation)', 'Total Provisions and Contingencies', 'Total Expenditure']
    for i in expenditure_cols:
        if i in df.columns:
            df[i] = df[i] / df['Total Assets'] * 100

    # normalizing net profit / loss for the Year items wrt total assets
    df['Net Profit / Loss for The Year'] = df['Net Profit / Loss for The Year'] / df['Total Assets']
    # normalizing assets wrt total assets
    assets_cols = ['Cash and Balances with Reserve Bank of India', 'Balances with Banks Money at Call and Short Notice', 'Advances', 'Investments', 'Fixed Assets', 'Other Assets']
    for i in assets_cols:
        if i in df.columns:
            df[i] = df[i] / df['Total Assets'] * 100
    # normalize  total reserves and surplus wrt total assets
    df['Total Reserves and Surplus'] = df['Total Reserves and Surplus'] / df['Total Assets']
    # normailize Total ShareHolders Funds Deposits, Borrowings, Other Liabilities and Provisions wrt Total Capital and Liabilities
    Funds_cols = ['Total ShareHolders Funds', 'Deposits', 'Borrowings', 'Other Liabilities and Provisions']
    for i in Funds_cols:
        if i in df.columns:
            df[i] = df[i] / df['Total Capital and Liabilities'] * 100
    # drop total capital and liabilities coloumn as it is 1 now
    df = df.drop('Total Capital and Liabilities', axis=1)
    # drop total assets coloumn as it is 1 now
    df = df.drop('Total Assets', axis=1)
    return df
```
